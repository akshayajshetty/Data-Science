# data preparation

```python
df.columns
```

```python
df = df[[#'Unnamed: 0', 
    'price', 'brand', 'model', 'year', 
    #'title_status',
       'mileage', 'color', #'vin', 
    #'lot', 
    'state', 'country', 
    #'condition'
]].copy()
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.rename(columns = {
    'price' : 'Price',
    'brand' : 'Brand',
    'model' : 'Model',
    'year' : 'Year',
    'mileage' : 'Mileage',
    'color' : 'Color',
    'state' : 'State',
    'country' : 'Country'
})
```

```python
df.head()
```

```python
df.drop('Mileage', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset = ['Brand'])].head()
```

```python
df.query("Brand== 'ford'")
```

```python
df = df.loc[~df.duplicated(subset=['Brand','Model','Year','State'])] \
    .reset_index(drop=True).copy()
```

```python
df.head()
```

```python
df.shape
```

# Feature understanding

```python
# bar chart
ax = df['Year'].value_counts().head(10).plot(kind='bar',title = 'Top 10 years Car Introduced') 
ax.set_xlabel('year introduced')
ax.set_ylabel('Count')
```

```python
ax = df['State'].value_counts().head(10).plot(kind='bar',title= 'Top states')
ax.set_xlabel('States')
```

```python
df['State'].value_counts()
```

```python
df.plot(kind='scatter', x='Color', y='Price', title = "States V/S Price")
plt.show()
```

```python
# pairplot
sns.pairplot(df,vars=['Price','Year'],hue = 'Model')
plt.show()
```

```python
df_corr = df[['Price','Year']].dropna().corr()
df_corr
```
