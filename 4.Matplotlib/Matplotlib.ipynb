# Axes, label, legend, grid

```python
days = ['Sun','Mon','Tues','Wed','Thurs','Fri','sat']
max_temp = [23,24,25,34,36,28,19]
min_temp = [20,16,17,18,20,18,15]
avg_temp = [21,20,22,25,27,23,16]
```

```python
plt.plot(days,max_temp)
plt.plot(days,min_temp)
plt.plot(days,avg_temp)
```

```python
plt.xlabel('days')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(days,max_temp)
plt.plot(days,min_temp)
plt.plot(days,avg_temp)      # Adding x , y and title labels
```

```python
plt.xlabel('days')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(days,max_temp,label='Max')
plt.plot(days,min_temp,label='Min')
plt.plot(days,avg_temp,label='Avg')
plt.legend(loc='upper left')              # Adding legend
```

```python
plt.xlabel('days')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(days,max_temp,label='Max')
plt.plot(days,min_temp,label='Min')
plt.plot(days,avg_temp,label='Avg')
plt.legend(loc='upper left',shadow=True) 
plt.grid()                       # Adding grid
```

# Work with Charts

# 1. Bar chart

```python
import numpy as np
```

```python
company = ["Google","Amazon","Microsoft","Salesforce","IBM"]
revenue = [25,8,10,9,14]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title('US stock')
plt.bar(xpos,revenue,label='revenue')
plt.legend()
```

```python
profit = [12,2,5,7,8]
```

```python
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title('US stock')
plt.bar(xpos,revenue,label='revenue')
plt.bar(xpos,profit,label='profit')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title('US stock')
plt.bar(xpos-0.2,revenue,label='revenue')
plt.bar(xpos+0.2,profit,label='profit')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title('US stock')
plt.bar(xpos-0.2,revenue,label='revenue',width=0.4)
plt.bar(xpos+0.2,profit,label='profit',width=0.4)
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title('US stock')
plt.barh(xpos,revenue,label='revenue')
plt.barh(xpos,profit,label='profit')
plt.legend()
```

# 2. Histogram

```python
mens_sugar_level = [80,85,81,90,95,98,100,107,96,103]
```

```python
plt.hist(mens_sugar_level)
```

```python
plt.hist(mens_sugar_level, bins=5)
```

```python
plt.hist(mens_sugar_level,bins=5,rwidth=0.80)
```

```python
plt.hist(mens_sugar_level,bins=[70,80,100,110,120,130],rwidth=0.80)
```

```python
mens_sugar_level = [80,85,81,90,95,98,100,107,96,103]
womens_sugar_level = [80,83,79,88,91,93,95,100,92,105]
```

```python
plt.hist([mens_sugar_level,womens_sugar_level],bins=[70,80,90,100,110,120],rwidth=0.90,color=['g','orange'],label=['men','women'])
```

```python
plt.hist([mens_sugar_level,womens_sugar_level],bins=[70,80,90,100,110,120],rwidth=0.90,color=['g','orange'],label=['men','women'])
plt.xlabel('Sugar_level')
plt.ylabel('No of patient')
plt.title('No of patient v/s sugar level')
plt.legend()
```

```python
plt.hist([mens_sugar_level,womens_sugar_level],bins=[70,80,90,100,110,120],rwidth=0.90,color=['g','orange'],label=['men','women'],orientation='horizontal')
plt.xlabel('Sugar_level')
plt.ylabel('No of patient')
plt.title('No of patient v/s sugar level')
plt.legend()                      # Horizonatal plot
```

# 3. Pie chart

```python
expense_value = [12000,1500,1000,2000,2000]
expense_label = ["Rent",'Food','Mobile bill','Petrol','Others']
```

```python
plt.pie(expense_value,labels=expense_label)
```

```python
plt.pie(expense_value,labels=expense_label)
plt.axis='equal'
```

```python
plt.pie(expense_value,labels=expense_label)
plt.axis='equal'
plt.show()
```

```python
plt.pie(expense_value,labels=expense_label,radius=0.8)
plt.axis='equal'
plt.show()
```

```python
plt.pie(expense_value,labels=expense_label,autopct='%0.1f%%')
plt.axis='equal'
plt.show()            # To conert to % (Percentage)
```

```python
plt.pie(expense_value,labels=expense_label,autopct='%0.1f%%',shadow=True)
plt.axis='equal'
plt.show() 
```

```python
plt.pie(expense_value,labels=expense_label,autopct='%0.1f%%',shadow=True,explode=[0,0.3,0,0,0])
plt.axis='equal'
plt.show()          # Remove one of the part
```

```python
plt.pie(expense_value,labels=expense_label,autopct='%0.1f%%',shadow=True,explode=[0,0.3,0,0,0],startangle=120)
plt.axis='equal'
plt.show()               # start the angle of first value
```

```python
plt.pie(expense_value,labels=expense_label,autopct='%0.1f%%',shadow=True,explode=[0,0.3,0,0,0],startangle=120,radius=0.8)
plt.axis='equal'
plt.show() 
plt.savefig("piechart.png",bbox_inches='tight')
```
