# 2. Concat

```python
india_weather = pd.DataFrame(
    {
        "city" : ['BNG','CKM','Mumbai'],
        "Temp" : [23,20,30],
        "Humidity" : [80,90,78]
        
    }
)
india_weather
```

```python
US_weather = pd.DataFrame(
    {
        "city" : ['NY','NJ','DC'],
        "Temp" : [23,20,30],
        "Humidity" : [80,90,78]
        
    }
)
US_weather
```

```python
df = pd.concat([india_weather,US_weather])
df
```

```python
df = pd.concat([india_weather,US_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,US_weather],keys = ["India","US"])
df
```

```python
df.loc["India"]
```

```python
df.loc["US"]
```

# 3. Merge

```python
Temp = pd.DataFrame(
    {
        "city" : ["BNG","CKM","Mumbai"],
        "Temp" : [25,20,28]
    }
)
Temp
```

```python
Humidity = pd.DataFrame(
    {
        "city" : ["BNG","CKM","Mumbai"],
        "Humidity" : [85,90,78]
    }
)
Humidity
```

```python
df = pd.merge(Temp,Humidity, on='city')
df
```

```python
Humidity_1 = pd.DataFrame(
    {
        "city" : ["BNG","CKM","Mumbai","Chennai"],
        "Humidity" : [85,90,78,70]
    }
)
Humidity_1
```

```python
df = pd.merge(Temp,Humidity_1, on='city')
df                           # Default inner join
```

```python
df = pd.merge(Temp,Humidity_1, on='city',how = 'left')
df                      # Left join
```

```python
df = pd.merge(Temp,Humidity_1, on='city', how = 'right')
df                 # Right join
```

```python
df = pd.merge(Temp,Humidity_1, on='city', how = 'outer')
df                 # Outer join
```

```python
df = pd.merge(Temp,Humidity_1, on='city',how = 'outer', indicator = True)
df
```
